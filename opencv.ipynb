```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image_bgr = cv2.imread('/content/11443977474_8096e6bbc3_k.jpg')
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2)
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
cropped = image_rgb[50:200, 100:300]
cv2.rectangle(image_bgr, (100, 50), (300, 200), (0, 255, 0), 2)
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("RGB Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(cropped)
plt.title("Cropped Image")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB))
plt.title("Image with Rectangle")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('sample.jpg')
cv2.putText(img, "Hello Students!", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 3, (0, 255, 0), 7)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Text on Image")
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('sample.jpg')
img_flip = cv2.flip(img, -1)  # 1 = Horizontal flip
img_rgb = cv2.cvtColor(img_flip, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Flipped Image")
plt.axis('off')
plt.show()
```
